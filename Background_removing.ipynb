## Multiple color backgroud

```python
filename_r='image/sprite_red.jpg'
filename_g='image/sprite_green.jpg'
filename_b='image/sprite_blue.jpg'

filename_r='image/bottle_red.jpg'
filename_g='image/bottle_green.jpg'
filename_b='image/bottle_blue.jpg'

filename_r='image/rino_red.jpg'
filename_g='image/rino_green.jpg'
filename_b='image/rino_blue.jpg'

ir = mpimg.imread(filename_r)
ig = mpimg.imread(filename_g)
ib = mpimg.imread(filename_b)

limit = 140

mask_r = np.where(ir[:,:,0]>150, 1,0) & np.where(ir[:,:,1]<limit, 1,0) & np.where(ir[:,:,2]<limit, 1,0)
mask_g = np.where(ig[:,:,0]<limit, 1,0) & np.where(ig[:,:,1]>150, 1,0) & np.where(ig[:,:,2]<limit, 1,0)
mask_b = np.where(ib[:,:,0]<limit, 1,0) & np.where(ib[:,:,1]<limit, 1,0) & np.where(ib[:,:,2]>150, 1,0)

mask_r = np.where(mask_r == 0, 1,0)
mask_g = np.where(mask_g == 0, 1,0)
mask_b = np.where(mask_b == 0, 1,0)

mask = (mask_r + mask_g + mask_b) >= 2

i = np.zeros((ir.shape), dtype='int')
i[:,:,0] = mask * ir[:,:,0] 
i[:,:,1] = mask * ir[:,:,1] 
i[:,:,2] = mask * ir[:,:,2] 

print("oui")

plt.figure(figsize=(20, 20))
plt.subplot(4,2,1)
plt.imshow(ir)
plt.subplot(4,2,2)
plt.imshow(mask_r, cmap='gray')
plt.subplot(4,2,3)
plt.imshow(ig)
plt.subplot(4,2,4)
plt.imshow(mask_g, cmap='gray')
plt.subplot(4,2,5)
plt.imshow(ib)
plt.subplot(4,2,6)
plt.imshow(mask_b, cmap='gray')
plt.subplot(4,2,7)
plt.imshow(i)
plt.subplot(4,2,8)
plt.imshow(mask, cmap='gray')
plt.subplots_adjust(wspace=-0.5, hspace=0.1)
plt.savefig('multipleBackgroundRemoving.png')
```
